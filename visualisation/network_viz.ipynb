# Network visualisation: LINNAE triples

```python
import glob
import networkx as nx
import netwulf as nw
import pandas as pd
import rdflib
```

```python
from rdflib.namespace import RDF, RDFS, OWL, XSD, FOAF, SKOS, DOAP, DC, DCTERMS, VOID

g = rdflib.Graph()
example = "1_2"
file = "https://raw.githubusercontent.com/LINNAE-project/SFB-Annotator/143e8105ffee826d23acd6a85a020aeff6091e80/data/rdf/example_"+example+".ttl"
g.parse(file, format='ttl')

ANNO = rdflib.Namespace("http://localhost:8080/rdf/nc/annotation/")
DWC = rdflib.Namespace("http://rs.tdwg.org/dwc/terms/")
NHC = rdflib.Namespace("http://makingsense.liacs.nl/rdf/nhc/")
DWCIRI =rdflib.Namespace("http://rs.tdwg.org/dwc/iri/")
OA =rdflib.Namespace("http://www.w3.org/ns/oa#")
DSW = rdflib.Namespace("http://purl.org/dsw/")
NC = rdflib.Namespace("http://makingsense.liacs.nl/rdf/nc/")
EX = rdflib.Namespace("http://localhost:8080/semanticAnnotator/files/")
DCMITYPE = rdflib.Namespace("http://purl.org/dc/dcmitype/")
g.bind("foaf", FOAF)
g.bind("rdf", RDF)
g.bind("dwc", DWC)
g.bind("dwciri", DWCIRI)
g.bind("dsw", DSW)
g.bind("nhc", NHC)
g.bind("dc", DC)
g.bind("oa", OA)
g.bind("dcterms", DCTERMS)
g.bind("nc", NC)
g.bind("ex",EX)
g.bind("dcmitype",DCMITYPE)
g.bind("anno",ANNO)
```

```python
print("graph has {} statements.".format(len(g)))
print(g.serialize(format="turtle").decode("utf-8"))
```

```python
G = nx.Graph()

#can't seem to use rdflib to extract a namespace from an iri, so used split(':') 
for subj, pred, obj in g:
    G.add_node(subj.n3(g.namespace_manager), group=subj.n3(g.namespace_manager).split(':')[0])
    G.add_node(obj.n3(g.namespace_manager), group=obj.n3(g.namespace_manager).split(':')[0])
    G.add_edge(subj.n3(g.namespace_manager),obj.n3(g.namespace_manager),label=pred.n3(g.namespace_manager))
    
network, config = nw.visualize(G)
```
